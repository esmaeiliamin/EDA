```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
#Set the aesthetic style of the plots
sns.set_style('whitegrid')
sns.set_style('white')
```

```python
#Generate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
#Normilize using Numpy
normilized_np = (data - np.min(data)) / (np.max(data) - np.min(data))

#Orginal Data
sns.histplot(data, kde=True)
```

```python
#Normilized Data with Numpy
sns.histplot(normilized_np, kde=True)
```

```python
#Normilize using scikit-learn
scaler =MinMaxScaler()
normilized_sklearn = scaler.fit_transform(data.reshape(-1, 1))
```

```python
#Normilized Data with sklearn
sns.histplot(normilized_sklearn, kde=True)
```

```python
#Standardization using Numpy
std_np = (data - np.mean(data)) / np.std(data)

sns.histplot(std_np, kde=True)
```

```python
from sklearn.preprocessing import StandardScaler
#Standardization using scikit-learn
scaler =StandardScaler()
std_skl = scaler.fit_transform(data.reshape(-1, 1))

sns.histplot(std_skl, kde=True)
```

```python
#Robust Scaling 
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
robust_scaler = scaler.fit_transform(data.reshape(-1, 1))

sns.histplot(robust_scaler, kde=True)
```
